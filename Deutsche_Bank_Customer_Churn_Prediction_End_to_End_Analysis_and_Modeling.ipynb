### Feature Selection

Feature selection is the process of identifying which variables to include in a predictive model. In practice, this happens at multiple stages of a data analytics project. While a data professional may sometimes receive a prepared dataset with a defined target, more often they start with a question or business problem. From there, they need to:

* Assess what data is available
* Define an appropriate target variable
* Decide on a modeling approach
* Assemble a set of features likely to be useful for predicting that target

This initial selection happens during the **planning stage**.

Feature selection is a continuous, layered process that happens both before and during model development. It begins during exploratory data analysis (EDA), where variables are assessed for appropriateness — for instance, identifying features with excessive null values or those unlikely to hold predictive value. Decisions like dropping irrelevant or problematic variables at this stage are also part of feature selection. 

Later, during modeling, it's common to build preliminary models, review feature importances or coefficients, and iteratively remove low-value predictors. This informal, model-based selection complements more structured techniques like Forward or Backward Elimination. Together, these steps help streamline models, improve performance, and focus on the most meaningful predictors.

### **Feature Extraction**

Beyond using existing variables, it’s often valuable to create new features derived from the data. Well-designed derived features can capture patterns or relationships not explicitly represented in the raw data, sometimes becoming some of the model’s most predictive inputs.

Effective feature extraction typically relies on a mix of domain knowledge and understanding of the dataset. For example, if a data professional knew that a system glitch caused many credit card transactions to fail in October, it might be reasonable to suspect those affected customers are more likely to churn. If the dataset contained monthly credit card transaction counts, one could engineer a feature like `OctUseRatio`:

$\text{OctUseRatio} = \frac{\text{Number of October transactions}}{\text{Average monthly transactions}}$

This new feature could help flag customers likely affected by the glitch and improve model performance. In this way, domain-aware feature creation can be a powerful tool for enriching a predictive model.

The dataset doesn’t include contextual or event-based features like transaction logs, which would allow for the creation of a metric such as `OctUseRatio`. However, I can still engineer a meaningful new feature from the existing data to potentially enhance the model’s predictive performance.

I’ll create a `Loyalty` feature that represents the percentage of each customer’s life that they were customers by dividing `Tenure` by `Age`:

$$
\text{Loyalty} = \frac{\text{Tenure}}{\text{Age}}
$$

The intuition here is that people who have been customers for a greater proportion of their lives might be less likely to churn.


```python
churn_df['Loyalty'] = churn_df['Tenure'] / churn_df['Age']
```

There are three unique values: France, Spain, and Germany. To encode this data for machine learning, I will use the `pd.get_dummies()` function, which will replace the `Geography` column with three new Boolean columns—one for each category.

When I specify `drop_first=True`, it will create only two new columns instead of three. This is because the third category (France) can be inferred if both `Geography_Germany` and `Geography_Spain` are 0, making the `Geography_France` column unnecessary. Dropping the first category simplifies the dataset and helps prevent multicollinearity. Multicollinearity can occur when features are highly correlated, which can distort model coefficients and make interpretation difficult. By dropping the first category, I avoid redundancy and ensure the model can interpret the data more effectively.

As a result, I will end up with two columns: `Geography_Germany` and `Geography_Spain`. If both are 0, I can infer the customer is from France.

```python
churn_df = pd.get_dummies(churn_df, drop_first=True, dtype=int)
```

Now, I'll examine more closely how my model classified the test data. I'll do this with a confusion matrix. A confusion matrix is a graphic that shows my model’s true and false positives, and true and false negatives. I can do this using the `ConfusionMatrixDisplay` and `confusion_matrix` functions that I imported.

Here’s a helper function that will allow me to plot a confusion matrix for my model.

```python
def conf_matrix_plot(model, x_data, y_data):
    '''
    Accepts as argument model object, X data (test or validate), and y data (test or validate). 
    Return a plot of confusion matrix for predictions on y data.
    ''' 
  
    model_pred = model.predict(x_data)
    cm = confusion_matrix(y_data, model_pred, labels=model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                  display_labels=model.classes_,
                                 )
  
    disp.plot(values_format='')
    
    plt.show()
```

Since the data imbalanced class distribution, I know that if I measure model performance by accuracy alone, the model could predict 0 (no churn) 100% of the time and still have an accuracy of \~80%. While an accuracy of 80% might initially seem good, in this case, it would be misleading because the model would fail to identify anyone who churned. Therefore, accuracy is not the best metric to evaluate the model's performance.

To determine which evaluation metric would be best, I need to consider how the model might be wrong. There are two possibilities for incorrect predictions:

* **False positives:** When the model predicts a customer **will** churn when in fact they won’t.
* **False negatives:** When the model predicts a customer will **not** churn when in fact they will.

Aside from accuracy, there are several other performance metrics to choose from, such as precision, recall, and F1 score. Here's a closer look at these metrics and how they apply to this case:

1. **Precision**
$$precision = \frac{\text{TP}}{\text{FP+TP}}$$
  </br> 

In my case, precision represents the percentage of all predicted positives by the model that are true positives. However, this metric might not be ideal because it disincentivizes predicting churn unless there’s a high level of certainty. This can lead to a higher number of false negatives, where I miss predicting churn for customers who will actually leave.

2. **Recall**:

$$recall = \frac{\text{TP}}{\text{FN+TP}}$$  
  </br>
  
Recall, on the other hand, represents the percentage of actual positives (churners) that the model identifies correctly. While this can help capture more churners, it might also result in a high number of false positives—predicting churn for customers who don’t actually leave.

For reference, one can refer to the confusion matrix to understand what each of these terms represents.

To decide which error is worse (false positives or false negatives), I first need to define what "worse" means in the context of this problem. In this business scenario, for the sake of this research question, let’s define "worse" as the error that results in the highest financial cost to the bank.

To assess this, I can get an idea of how much money each customer who churns costs the bank by calculating the average balance of all customers who churned. This will give me insight into the financial impact of false positives versus false negatives.

```python
avg_churned_bal = churn_df1[churn_df1['Exited']==1]['Balance'].mean()
avg_churned_bal
```

This shows that the customers who churned each took with them **€91,108.54**, on average. That's a lot of money! This represents the average cost of the model predicting a false negative.

As for the cost of predicting a false positive, it's the cost of whatever incentive might be offered to convince someone to stay with the bank when they were going to stay regardless. I don’t have an exact number for this, but even though it’s probably less than **€91,108.54**, it could still be thousands of euros per customer in lost revenue, depending on the details of the incentive.

Since correctly identifying customers who will churn is potentially very valuable, I could select recall as my most important metric. This might be a perfectly valid approach, depending on the specifics of the campaign. However, this could also be problematic. After all, if I select a model based solely on recall, I could end up with a very biased model that predicts everyone will churn, leading to 8,000 people being given incentives unnecessarily.

Since I don’t know the exact cost of predicting a false negative, I’ll make an assumption for this exercise. I’ll assume that a metric that balances precision and recall is best. The metric that helps achieve this balance is the *F1 score*, which is defined as the harmonic mean of precision and recall.

$${F_{1}} = 2 \cdot \frac{precision \cdot  recall}{precision + recall}$$

Again, there are many metrics to choose from. The key is to make an informed decision based on the specific business case and research question at hand.

As a reminder, when I ran the grid search, I also specified that I wanted to capture precision, recall, and accuracy in addition to the F1 score. The reason for doing this is that it's challenging to interpret an F1 score on its own. These other metrics—precision, recall, and accuracy—are more directly interpretable, making them important to consider.

Next, a helper function is defined to extract these metrics from the fitted `GridSearchCV` object and return a pandas DataFrame containing all four scores (precision, recall, accuracy, and F1) from the model that achieved the best average F1 score during validation. This function will be useful later when adding the results of other models to the comparison table.

```python
def make_results(model_name, model_object):
    '''
    Accepts as arguments a model name (a string of any choice) and a fit GridSearchCV model object.
  
    Returns a pandas df with the F1, recall, precision, and accuracy scores
    for the model with the best mean F1 score across all validation folds.  
    '''

    # Get all the results from the CV and put them in a df
    cv_results = pd.DataFrame(model_object.cv_results_)

    # Isolate the row of the df with the max(mean f1 score)
    best_estimator_results = cv_results.iloc[cv_results['mean_test_f1'].idxmax(), :]

    # Extract accuracy, precision, recall, and f1 score from that row
    f1 = best_estimator_results.mean_test_f1
    recall = best_estimator_results.mean_test_recall
    precision = best_estimator_results.mean_test_precision
    accuracy = best_estimator_results.mean_test_accuracy
  
    # Create table of results
    table = pd.DataFrame()
    # Create table of results
    table = pd.DataFrame({'Model': [model_name],
                          'F1': [f1],
                          'Recall': [recall],
                          'Precision': [precision],
                          'Accuracy': [accuracy]
                         }
                        )
  
    return table
```

### Hyperparameters tuned with separate validation set

I’ll begin by splitting the training data to create a validation dataset. I won’t touch the test data at all.

I’ll use `train_test_split` to divide `X_train` and `y_train` into 80% training data (`X_tr`, `y_tr`) and 20% validation data (`X_val`, `y_val`). I won’t forget to stratify it and set the random state.

```python
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.2, 
                                            stratify=y_train, random_state=10)
```

When I tune hyperparameters with `GridSearchCV` using a separate validation dataset, I have to take a few extra steps. `GridSearchCV` by default cross-validates the data. In fact, if I left the `cv` argument blank, it would split the data into five folds for cross-validation by default.

I don’t want it to do this. Instead, I’ll tell it exactly which rows of `X_train` are for training, and which rows are for validation.

To do this, I need to make a list of length `len(X_train)` where each element is either a 0 or -1. A 0 in index *i* will indicate to `GridSearchCV` that index *i* of `X_train` is to be held out for validation. A -1 at a given index will indicate that that index of `X_train` is to be used as training data.

I’ll make this list using a list comprehension that looks at the index number of each row in `X_train`. If that index number is in `X_val`'s list of index numbers, then the list comprehension appends a 0. If it’s not, then it appends a -1.

So if the training data is:
\[A, B, C, D],
and the list is:
\[-1, 0, 0, -1],
then `GridSearchCV` will use a training set of \[A, D] and a validation set of \[B, C].


```python
split_index = [0 if x in X_val.index else -1 for x in X_train.index]
```

Now I can build the model. Everything stays the same as when I was cross-validating earlier — except this time, I pass the `split_index` list to the `PredefinedSplit` function and assign the result to a new variable called `custom_split`.

Then, I’ll use this `custom_split` for the `cv` argument when I instantiate `GridSearchCV`.

```python
base_random_forest1 = RandomForestClassifier(random_state=0)

cv_params = {
    'max_depth': [2, 3, 4, 5, None], 
    'min_samples_leaf': [1, 2, 3],
    'min_samples_split': [2, 3, 4],
    'max_features': [2, 3, 4],
    'n_estimators': [75, 100, 125, 150]
}  

scoring = ["accuracy", "precision", "recall", "f1"]

custom_split = PredefinedSplit(split_index)

tuned_random_forest_csval = GridSearchCV(base_random_forest1, cv_params, scoring=scoring, cv=custom_split, refit="f1")
```

## A Reminder About Modeling Trade-offs

The decision to use only the champion model to predict on the test data comes with a trade-off. The benefit is that it provides a true sense of how I’d expect the model to perform on new, unseen data. The cost, however, is that by using validation scores to both tune hyperparameters **and** select the champion model, there is a risk of choosing the model that happened to perform best on the validation data by chance, which could potentially be one that overfit to it.

An alternative would have been to select the champion model by having all tuned models predict on the test data and choosing the best performer. But this too comes with a trade-off. While it reduces the risk of overfitting to the validation set, it compromises the objectivity of the test set as a final, unbiased check of model performance. In this case, a fresh dataset would be needed for a true final evaluation, which would have required setting more data aside upfront and leaving less for training.

A more rigorous workflow, when sufficient data is available, would be:

1. Split the data into training, validation, and test sets.
2. Tune hyperparameters using cross-validation on the training set.
3. Use *all* tuned models to predict on the validation set.
4. Select a champion model based on validation performance.
5. Use the champion model alone to predict on the test set.

Every modeling decision involves trade-offs. What matters most is recognizing them and applying the best reasoning suited to the problem, dataset size, and business context.
