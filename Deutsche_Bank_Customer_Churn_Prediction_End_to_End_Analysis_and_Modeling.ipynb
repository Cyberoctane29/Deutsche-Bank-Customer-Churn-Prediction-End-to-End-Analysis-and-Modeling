### Feature Selection

Feature selection is the process of identifying which variables to include in a predictive model. In practice, this happens at multiple stages of a data analytics project. While a data professional may sometimes receive a prepared dataset with a defined target, more often they start with a question or business problem. From there, they need to:

* Assess what data is available
* Define an appropriate target variable
* Decide on a modeling approach
* Assemble a set of features likely to be useful for predicting that target

This initial selection happens during the **planning stage**.

Feature selection is a continuous, layered process that happens both before and during model development. It begins during exploratory data analysis (EDA), where variables are assessed for appropriateness — for instance, identifying features with excessive null values or those unlikely to hold predictive value. Decisions like dropping irrelevant or problematic variables at this stage are also part of feature selection. 
Later, during modeling, it's common to build preliminary models, review feature importances or coefficients, and iteratively remove low-value predictors. This informal, model-based selection complements more structured techniques like Forward or Backward Elimination. Together, these steps help streamline models, improve performance, and focus on the most meaningful predictors.

### **Feature Extraction**

Beyond using existing variables, it’s often valuable to create new features derived from the data. Well-designed derived features can capture patterns or relationships not explicitly represented in the raw data, sometimes becoming some of the model’s most predictive inputs.

Effective feature extraction typically relies on a mix of domain knowledge and understanding of the dataset. For example, if a data professional knew that a system glitch caused many credit card transactions to fail in October, it might be reasonable to suspect those affected customers are more likely to churn. If the dataset contained monthly credit card transaction counts, one could engineer a feature like `OctUseRatio`:

$\text{OctUseRatio} = \frac{\text{Number of October transactions}}{\text{Average monthly transactions}}$

This new feature could help flag customers likely affected by the glitch and improve model performance. In this way, domain-aware feature creation can be a powerful tool for enriching a predictive model.

The dataset doesn’t include contextual or event-based features like transaction logs, which would allow for the creation of a metric such as `OctUseRatio`. However, I can still engineer a meaningful new feature from the existing data to potentially enhance the model’s predictive performance.

I’ll create a `Loyalty` feature that represents the percentage of each customer’s life that they were customers by dividing `Tenure` by `Age`:

$$
\text{Loyalty} = \frac{\text{Tenure}}{\text{Age}}
$$

The intuition here is that people who have been customers for a greater proportion of their lives might be less likely to churn.


```python
churn_df['Loyalty'] = churn_df['Tenure'] / churn_df['Age']
```

There are three unique values: France, Spain, and Germany. To encode this data for machine learning, I will use the `pd.get_dummies()` function, which will replace the `Geography` column with three new Boolean columns—one for each category.

When I specify `drop_first=True`, it will create only two new columns instead of three. This is because the third category (France) can be inferred if both `Geography_Germany` and `Geography_Spain` are 0, making the `Geography_France` column unnecessary. Dropping the first category simplifies the dataset and helps prevent multicollinearity. Multicollinearity can occur when features are highly correlated, which can distort model coefficients and make interpretation difficult. By dropping the first category, I avoid redundancy and ensure the model can interpret the data more effectively.

As a result, I will end up with two columns: `Geography_Germany` and `Geography_Spain`. If both are 0, we can infer the customer is from France.

```python
churn_df = pd.get_dummies(churn_df, drop_first=True, dtype=int)
```

```python
def write_pickle(path, model_object, save_as:str):
    '''
    In:
        path:         folder path where the model should be saved  
        model_object: the trained model to pickle  
        save_as:      filename to save the model as (with .pkl extension)

    Out: Saves the model as a pickle file in the specified folder
    '''


    with open(path + save_as + '.pickle', 'wb') as to_write:
        pickle.dump(model_object, to_write)
```

```python
def read_pickle(path, saved_model_name:str):
    '''
    In:  
        path:             folder path to load the model from  
        saved_model_name: filename of the pickled model to load  

    Out:  
        model: the unpickled model object  
    '''

    with open(path + saved_model_name + '.pickle', 'rb') as to_read:
        model = pickle.load(to_read)

    return model
```

As noted earlier, the class of my target variable is split roughly 80/20. In other words, ~20% of the people in this dataset churned. This is an imbalanced dataset, but it’s not extreme. I will preserve this ratio when I model.

Since the data imbalanced class distribution, I know that if I measure model performance by accuracy alone, the model could predict 0 (no churn) 100% of the time and still have an accuracy of \~80%. While an accuracy of 80% might initially seem good, in this case, it would be misleading because the model would fail to identify anyone who churned. Therefore, accuracy is not the best metric to evaluate the model's performance.

To determine which evaluation metric would be best, I need to consider how the model might be wrong. There are two possibilities for incorrect predictions:

* **False positives:** When the model predicts a customer **will** churn when in fact they won’t.
* **False negatives:** When the model predicts a customer will **not** churn when in fact they will.

Aside from accuracy, there are several other performance metrics to choose from, such as precision, recall, and F1 score. Here's a closer look at these metrics and how they apply to this case:

1. **Precision**
$$precision = \frac{\text{TP}}{\text{FP+TP}}$$
  </br> 

In my case, precision represents the percentage of all predicted positives by the model that are true positives. However, this metric might not be ideal because it disincentivizes predicting churn unless there’s a high level of certainty. This can lead to a higher number of false negatives, where I miss predicting churn for customers who will actually leave.

2. **Recall**:

$$recall = \frac{\text{TP}}{\text{FN+TP}}$$  
  </br>
  
Recall, on the other hand, represents the percentage of actual positives (churners) that the model identifies correctly. While this can help capture more churners, it might also result in a high number of false positives—predicting churn for customers who don’t actually leave.

For reference, one can refer to the confusion matrix to understand what each of these terms represents.

To decide which error is worse (false positives or false negatives), I first need to define what "worse" means in the context of this problem. In this business scenario, for the sake of this research question, let’s define "worse" as the error that results in the highest financial cost to the bank.

As a reminder, when I ran the grid search, I also specified that I wanted to capture precision, recall, and accuracy in addition to the F1 score. The reason for doing this is that it's challenging to interpret an F1 score on its own. These other metrics—precision, recall, and accuracy—are more directly interpretable, making them important to consider.

The next step is defining a helper function that extracts these metrics from the fitted `GridSearchCV` object and returns a pandas dataframe containing all four scores (precision, recall, accuracy, and F1) from the model that achieved the best average F1 score during validation. This function will be useful later when adding the results of other models to the comparison table.

```python
def make_results(model_name, model_object):
    '''
    Accepts as arguments a model name (your choice - string) and
    a fit GridSearchCV model object.
  
    Returns a pandas df with the F1, recall, precision, and accuracy scores
    for the model with the best mean F1 score across all validation folds.  
    '''

    # Get all the results from the CV and put them in a df
    cv_results = pd.DataFrame(model_object.cv_results_)

    # Isolate the row of the df with the max(mean f1 score)
    best_estimator_results = cv_results.iloc[cv_results['mean_test_f1'].idxmax(), :]

    # Extract accuracy, precision, recall, and f1 score from that row
    f1 = best_estimator_results.mean_test_f1
    recall = best_estimator_results.mean_test_recall
    precision = best_estimator_results.mean_test_precision
    accuracy = best_estimator_results.mean_test_accuracy
  
    # Create table of results
    table = pd.DataFrame()
    # Create table of results
    table = pd.DataFrame({'Model': [model_name],
                          'F1': [f1],
                          'Recall': [recall],
                          'Precision': [precision],
                          'Accuracy': [accuracy]
                         }
                        )
  
    return table
```
